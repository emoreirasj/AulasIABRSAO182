boto3: A biblioteca oficial da AWS para Python, para criar um cliente que se comunica com o serviço Amazon Bedrock e consome o modelo

import boto3

def get_bedrock_client():
    return boto3.client(
        "bedrock-runtime",
        aws_access_key_id=AWS_ACCESS_KEY_ID,
        aws_secret_access_key=AWS_SECRET_ACCESS_KEY,
        region_name=AWS_REGION
    )


def invoke_model(prompt):
    client = get_bedrock_client()
    response = client.invoke_model(
        modelId="amazon.titan-tg1-large",
        contentType="application/json",
        accept="application/json",
        body='{"prompt": "' + prompt + '"}'
    )
    return response["body"].read().decode("utf-8")



boto3.client("bedrock-runtime") → Inicializa um cliente AWS para se conectar ao serviço Amazon Bedrock.

Autenticação: Passa as credenciais de acesso (AWS_ACCESS_KEY_ID, AWS_SECRET_ACCESS_KEY) e define a região (AWS_REGION).

Essa função retorna um objeto cliente que pode ser usado para enviar requisições ao Amazon Bedrock.

client.invoke_model(...) → Faz uma chamada ao modelo Titan, passando os seguintes parâmetros:

modelId="amazon.titan-tg1-large" → Identificador do modelo Titan a ser utilizado.

contentType="application/json" → Define que o conteúdo da requisição é JSON.

accept="application/json" → Define que a resposta será em formato JSON.

body='{"prompt": "' + prompt + '"}' → Envia o texto digitado pelo usuário para processamento pelo modelo.
